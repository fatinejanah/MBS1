```python
#Importer lespackages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#Lire la base de donn√©es
bd=pd.read_csv('/LoanExport.csv')
bd
```

```python
#1. Data Cleaning:
#Identify
bd.isnull().sum()
```

```python
#  handle missing values
bd.dropna(inplace=True)
bd
```

```python
# Remove duplicates
bd.drop_duplicates(inplace=True)
bd
```

```python
#Correct data types :
# Conversion des colonnes num√©riques
bd['CreditScore'] = bd['CreditScore'].astype('int')
bd['FirstPaymentDate'] = pd.to_datetime(bd['FirstPaymentDate'], format='%Y%m')
bd['MaturityDate'] = pd.to_datetime(bd['MaturityDate'], format='%Y%m')
bd['OrigUPB'] = bd['OrigUPB'].astype('int')
bd['OrigInterestRate'] = bd['OrigInterestRate'].astype('float')

# Conversion des colonnes cat√©gorielles
bd['FirstTimeHomebuyer'] = bd['FirstTimeHomebuyer'].astype('category')
bd['Channel'] = bd['Channel'].astype('category')
bd['ProductType'] = bd['ProductType'].astype('category')
bd['PropertyState'] = bd['PropertyState'].astype('category')
bd['PropertyType'] = bd['PropertyType'].astype('category')
bd['SellerName'] = bd['SellerName'].astype('category')
bd['ServicerName'] = bd['ServicerName'].astype('category')
bd
```

```python
#2. Data Encoding:
le = LabelEncoder()
bd['FirstTimeHomebuyer'] = le.fit_transform(bd['FirstTimeHomebuyer'])
bd['Channel'] = le.fit_transform(bd['Channel'])
bd['ProductType'] = le.fit_transform(bd['ProductType'])
bd['PropertyState'] = le.fit_transform(bd['PropertyState'])
bd['PropertyType'] = le.fit_transform(bd['PropertyType'])
bd['SellerName'] = le.fit_transform(bd['SellerName'])
bd['ServicerName'] = le.fit_transform(bd['ServicerName'])
bd
```

```python
#3. Data Labelling:
# S√©parer les caract√©ristiques et la cible
target = bd['EverDelinquent']
features = bd.drop('EverDelinquent', axis=1)
bd
```

```python
# Enregistrer les donn√©es pr√©trait√©es dans un nouveau fichier CSV
bd.to_csv('LoantExport.csv', index=False)
```
